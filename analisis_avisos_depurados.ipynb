```python
# Modificar formatos de fecha
df_avisos[' Fecha inicio'] = pd.to_datetime(df_avisos[' Fecha inicio'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Fecha de fin'] = pd.to_datetime(df_avisos[' Fecha de fin'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Modificado el'] = pd.to_datetime(df_avisos[' Modificado el'], format='%Y-%m-%d', errors='coerce')
df_avisos['Inicio de averia'] = pd.to_datetime(df_avisos['Inicio de averia'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Fin de averia'] = pd.to_datetime(df_avisos[' Fin de averia'], format='%Y-%m-%d', errors='coerce')
df_avisos['Creado el'] = pd.to_datetime(df_avisos['Creado el'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Fin deseado'] = pd.to_datetime(df_avisos[' Fin deseado'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Cierre por fecha'] = pd.to_datetime(df_avisos[' Cierre por fecha'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Fecha de aviso'] = pd.to_datetime(df_avisos[' Fecha de aviso'], format='%Y-%m-%d', errors='coerce')
df_avisos[' Inicio deseado'] = pd.to_datetime(df_avisos[' Inicio deseado'], format='%Y-%m-%d', errors='coerce')
```

```python
df_avisos = df_avisos.dropna(subset=[' Clase de aviso'])
```

```python
# Gráfica clase de avisos
clases_avisos = df_avisos[' Clase de aviso'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=clases_avisos.index, y=clases_avisos.values, palette='viridis', hue=clases_avisos.index)
plt.xticks(rotation=45)
plt.xlabel('Clase de aviso')
plt.ylabel('Número de avisos')
plt.title('Clases de aviso')
plt.tight_layout()
plt.show()
```

```python
# Gráfica de las 30 ubicaciones técnicas más frecuentes (Con clase de aviso)
top_30_ubicaciones = df_avisos.groupby([' Ubicacion tecnica', ' Clase de aviso']).size().nlargest(30).reset_index(name='Frecuencia')

# print(top_30_ubicaciones)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_30_ubicaciones, x=' Ubicacion tecnica', y='Frecuencia', hue=' Clase de aviso', palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Ubicación técnica')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Clases de Avisos por Ubicación Técnica (Top 20)')
plt.legend(title='Clase de Aviso', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Gráfica de las primeras 10 áreas encargadas
area_encargada = df_avisos[' Perfil de catalogo'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=area_encargada.index, y=area_encargada.values, palette='viridis', hue=area_encargada.index)
plt.xticks(rotation=45)
plt.xlabel('Área encargada')
plt.ylabel('Número de avisos')
plt.title('Áreas encargadas')
plt.tight_layout()
plt.show()
```

```python
# Limpieza de datos (Coordenadas)
df_coordenadas = df_avisos[['Coordenada x;;', 'Coordenada y']].dropna()

# Limpieza de coordenada x (Cambio de formato)
df_coordenadas["Coordenada x"] = df_coordenadas["Coordenada x;;"].str.replace(';', '', regex=False).astype(float)
df_coordenadas.drop(columns=['Coordenada x;;'], inplace=True)
df_coordenadas.dropna(inplace=True)

# Eliminación de coordenadas con valores cero
df_coordenadas = df_coordenadas[(df_coordenadas['Coordenada x'] != 0) & (df_coordenadas['Coordenada y'] != 0)]
```

```python
# Gráfica con fechas de aviso
plt.figure(figsize=(12, 6))
sns.histplot(df_avisos[' Fecha de aviso'], bins=30, kde=True, color='blue')
plt.xlabel('Fecha de aviso')
plt.ylabel('Número de avisos')
plt.title('Distribución de Avisos por Fecha')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfica de clases de aviso por fecha
fecha_avisos = df_avisos.groupby([' Fecha de aviso', ' Clase de aviso']).size().reset_index(name='Frecuencia')
fecha_avisos

plt.figure(figsize=(12, 6))
sns.lineplot(data=fecha_avisos, x=' Fecha de aviso', y='Frecuencia', hue=' Clase de aviso', marker='o')
plt.xlabel('Fecha de aviso')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Clases de Avisos por Fecha')
plt.legend(title='Clase de Aviso', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Gráfica de estado del sistema 
status_counts = df_avisos['Status del sistema'].value_counts().iloc[2:21]

print(df_avisos['Status del sistema'].value_counts().iloc[:1])

plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='viridis', hue=status_counts.index)
plt.xlabel('Estado del sistema')
plt.ylabel('Número de avisos')
plt.title('Estado del sistema')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Gráfica de estado del sistema por fecha
status_fecha = df_avisos.groupby([' Fecha de aviso', 'Status del sistema']).size().reset_index(name='Frecuencia')

plt.figure(figsize=(12, 10))
sns.lineplot(data=status_fecha, x=' Fecha de aviso', y='Frecuencia', hue='Status del sistema', marker='o')
plt.xlabel('Fecha de aviso')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Estado del Sistema por Fecha')
plt.legend(title='Estado del Sistema', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Gráfica de poblaciones
poblacion = df_avisos[' Poblacion'].value_counts().head(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=poblacion.index, y=poblacion.values, palette='viridis', hue=poblacion.index)
plt.xticks(rotation=90)
plt.xlabel('Población')
plt.ylabel('Número de avisos')
plt.title('Población')
plt.tight_layout()
plt.show()
```
